```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\Tejaswi.k\OneDrive\Desktop\datascience coaching\ml\csv files\airlines_customer_satisfaction.csv")
```

```python
df.isnull().sum()
```

```python
# # Remove unnecessary dataset
df.drop(['Unnamed: 0', 'id'],axis=1, inplace=True)
```

```python
##apply median
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(
    df['Arrival Delay in Minutes'].median()
)
```

```python
##label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
##Split Features and Target
x = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05, random_state=7)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB(var_smoothing=0.001)
clf.fit(x_train,y_train)
```

```python
#train predictions
train_predictions = clf.predict(x_train)

#test predictions
test_predictions = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# train accuracy
print("Train accuracy: ",accuracy_score(y_train, train_predictions) * 100)

# test accuracy
print("Test accuracy: ",accuracy_score(y_test, test_predictions) * 100)
```

```python
print(confusion_matrix(y_test, test_predictions))
```

```python
print(classification_report(y_test, test_predictions))
```
